### Question 1: <b>10 points</b>
<div class="alert alert-block alert-success">
* You must write a function that uses the modulus oprator to determine if a given year is a leap year.
* The function is defined as:  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<code><b>int IsLeapYear(int year);</b></code>


* Where the parameter <b>year</b> is the year to check. 
* The function must return <b>1</b> if the year is a leap year and <b>0</b> if it is not.
* The algorithm for determining if a year is a leap year is:<b><br>
    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if (year is not divisible by 4) then (it is a common year)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else if (year is not divisible by 100) then (it is a leap year)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else if (year is not divisible by 400) then (it is a common year)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;else (it is a leap year)</b>

* The function will be tested automatically when you run the cell.
* Use the template below and fill in the code for <b>IsLeapYear()</b>
    </div>

```python
#include <stdio.h>
#include <string.h>
#include <stdlib.h>

/*****************************************************
 *
 *  Function:  IsLeapYear()
 *
******************************************************/
int IsLeapYear(int year)
{
    int result;
    
    /* Student Solution Goes Here */
    result = (year % 4) ? 0 : ((year % 100) ? 1 : (year % 400) ? 0 : 1);
                               
    return result;
}


/*****************************************************
 *
 *  Function:  main()
 *
******************************************************/
int main(int argc, char** argv)
{
    int yearsToTest[8] = { 1977, 1600, 1858, 1602, 2008, 2000, 2002, 1708 };
    
    for (int i = 0; i < 8; ++i)
    {
        printf("The year %d is %sa leap year\n", yearsToTest[i], 
               (IsLeapYear(yearsToTest[i]) == 1) ? "" : "not ");
    }
    return 0;
}
```

* Write a function that reads input as a stream of characters using fgetc().
* The function must continue reading until the tilde character (~) is encountered.
* The function must count and return the total number of words in the input.
* The function must also calculate and return the average number of characters per word.
* The function is defined as follows:<br><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<code><b>int ReadAndCountWords(int *pWordCount, float *pAvgChars);</b></code>


* Where the parameter <b>pWordCount</b> is used to return the number of words, and
* the parameter <b>pAvgChars</b> is used to return the average characters per word.

* The function will be tested automatically when you run the cell.

</div>


```python
#include <stdio.h>
#include <string.h>
#include <stdlib.h>

/*****************************************************
 *
 *  Function:  ReadAndCountWords()
 *
******************************************************/
int ReadAndCountWords(int *pWordCount, float *pAvgChars)
{
    int ch;
    int charCount=0;
    int curWordCount=0;
    int wordCount=0;
    while (1)  // forever, or at least until the loop breaks.
    {
        // this only returned after the user hits enter, but characters 
        // in the input buffer are returned one by one.
        ch = fgetc(stdin);

        // if EOF is encountered, print message and break the while() loop
        if (ch == '~')
        {
            printf("Exiting\n");
            break;
        }
        
        // these are known word separators
        if (ch == ' ' || ch == '\n' || ch == '\t')
        {
            if (curWordCount > 0)
            {
                curWordCount = 0;
                wordCount++;
            }
        }
        else if (!ispunct(ch))
        {
            ++curWordCount;
            ++charCount;
        }
    }
    *pWordCount = wordCount;
    *pAvgChars = (float)charCount/wordCount;
    
    return 0;
}

/*****************************************************
 *
 *  Function:  main()
 *
 *                   DO NOT MODIFY
******************************************************/
int main(int argc, char** argv)
{
    int wordCount;
    float averageCharsPerWord;

    ReadAndCountWords(&wordCount, &averageCharsPerWord);

    printf("Total number of words counted: %d\n", wordCount);
    if (wordCount > 0)
    {
        printf("Average number of characters per word: %0.2f\n", averageCharsPerWord);
    }
    return 0;
}
```
